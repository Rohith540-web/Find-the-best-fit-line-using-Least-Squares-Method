```python
print("Name: ROHITH V")
print("Reg no: 212224220083")
print("X and Y values")

import numpy as np
import matplotlib.pyplot as plt
x=np.array(eval(input()))
y=np.array(eval(input()))
```

```python
mean1=np.mean(x)
mean2=np.mean(y)
num=0
denom=0
for i in range(len(x)):
    num+=(x[i]-mean2)*(y[i]-mean2)
    denom+=(x[i]-mean1)**2
    
m=num/denom
b=mean2-m*mean1
print("slope:",m)
print("y_intercept:",b)
y_predicted=m*x+b
print("y predicted:",y_predicted)
```

```python
plt.scatter(x,y)
plt.plot(x,y_predicted,color='red')
print("Graph")
plt.show()

```
